```python
import pandas as pd
```

```python
# Read the CSV file
df = pd.read_csv('recent.csv', header=0)
df.head()
```

```python
df['Topics'] = df['Topics'].apply(lambda x: x.split(',') if isinstance(x, str) else [])
# strip whitespace from each topic
df['Topics'] = df['Topics'].apply(lambda x: [topic.strip() for topic in x])
df['Companies'] = df['Companies'].apply(lambda x: x.split(',') if isinstance(x, str) else [])
# strip whitespace from each company
df['Companies'] = df['Companies'].apply(lambda x: [company.strip() for company in x])
```

```python
df.head()
```

```python
# generate a pie chart of the top 10 most common values in the 'Topics' column (which is a list)
topics = df['Topics'].explode().value_counts().head(10)
topics.plot.pie(autopct='%1.1f%%', figsize=(10, 6), title='Top 10 Topics')
```

```python
# 10 most common values in the 'Companies' column (which is a list)
companies = df['Companies'].explode().value_counts().head(10)
companies.plot.pie(autopct='%1.1f%%', figsize=(10, 6), title='Top 10 Companies')
```

```python
# 10 most common Topics from 'Amazon' company
amazon_topics = df[df['Companies'].apply(lambda x: 'Amazon' in x)]['Topics'].explode().value_counts().head(10)
amazon_topics.plot.pie(autopct='%1.1f%%', figsize=(10, 6), title='Top 10 Topics for Amazon')
```

```python
# 10 most common Topics from 'Amazon' company
amazon_topics = df[df['Companies'].apply(lambda x: 'Google' in x)]['Topics'].explode().value_counts().head(10)
amazon_topics.plot.pie(autopct='%1.1f%%', figsize=(10, 6), title='Top 10 Topics for Google')
```
